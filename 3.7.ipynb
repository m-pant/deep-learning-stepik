```python
# 6 Basic Block Example
import torch
import torch.nn as nn

# здесь объявляйте класс модели

batch_size = 8
x = torch.rand(batch_size, 64, 32, 32) # тензор x в программе не менять

# здесь продолжайте программу
class BasicBlock(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.basic_block = nn.Sequential(
            nn.Conv2d(64, 64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
        )

    def forward(self, x):
        f = self.basic_block(x)
        return f + x

model_bb = BasicBlock()
model_bb.eval()
y = model_bb(x)

```

```python
# 7 Bottleneck Block Example

import torch
import torch.nn as nn

# здесь объявляйте класс модели

batch_size = 4
x = torch.rand(batch_size, 256, 16, 16) # тензор x в программе не менять

# здесь продолжайте программу
class BottleNeckBlock(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.bottleneck_block = nn.Sequential(
            nn.Conv2d(256, 64, 1, padding=0, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 256, 1, padding=0, bias=False),
            nn.BatchNorm2d(256),
        )

    def forward(self, x):
        f = self.bottleneck_block(x)
        return (f+x).relu()
    
model_bn = BottleNeckBlock()
model_bn.eval()
y = model_bn(x)
print(model_bn)
```

```python
# 7 Bottleneck Block Example v2
import torch
import torch.nn as nn

# здесь объявляйте класс модели

batch_size = 4
x = torch.rand(batch_size, 128, 16, 16) # тензор x в программе не менять

# здесь продолжайте программу

# здесь продолжайте программу
class BottleNeckBlock(nn.Module):
    def __init__(self, *args, **kwargs):
        super().__init__(*args, **kwargs)
        self.bottleneck_block = nn.Sequential(
            nn.Conv2d(128, 64, 1, padding=0, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 64, 3, padding=1, bias=False),
            nn.BatchNorm2d(64),
            nn.ReLU(inplace=True),
            nn.Conv2d(64, 256, 1, padding=0, bias=False),
            nn.BatchNorm2d(256),
        )

        self.skip_conn = nn.Sequential(
            nn.Conv2d(128, 256, 1, padding=0, bias=False),
            nn.BatchNorm2d(256)
        )

    def forward(self, x):
        f = self.bottleneck_block(x)
        sk = self.skip_conn(x)
        return (f+sk).relu()
    
model_bn = BottleNeckBlock()
model_bn.eval()
y = model_bn(x)
print(model_bn)
```
